```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('/content/train.csv')
```

```python
train.head()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

```python
train.shape
```

```python
train = train.drop(columns='Cabin', axis = 1)
train = train.drop(columns='Name', axis = 1)
```

```python
train = train.drop(columns= 'Embarked', axis = 1)
```

```python
train = train.drop(columns= 'Ticket', axis = 1)
```

```python
train['Fare'] = train['Fare']/513
```

```python
train.isnull().sum()
```

```python
train.tail()
```

```python
train['Sex'].replace({'male': 1 }, inplace = True)
```

```python
train['Sex'].replace({'female': 0 }, inplace = True)
```

```python
train.head()
```

```python
train_without_missing = train.dropna()
```

```python
train_without_missing.isnull().sum()
```

```python
train_without_missing.shape
```

```python
X = train_without_missing.drop('Survived', axis =1)
```

```python
Y = train_without_missing['Survived']
```

```python
X.head()
```

```python
Y.head()
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2, stratify = Y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
classifier = LogisticRegression()
```

```python
train.corr()
```

```python
classifier.fit(X_train, Y_train)
```

```python
Y_train_pred = classifier.predict(X_train)
```

```python
accuracy_train = accuracy_score(Y_train_pred, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(Y_pred, Y_test)
```
